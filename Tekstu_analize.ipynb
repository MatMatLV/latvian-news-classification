# Datu ielāde

```python
import pandas as pd
raksti_sports = pd.read_json('extracts/sports.json')
raksti_politika = pd.read_json('extracts/politika.json')
raksti_literatura = pd.read_json('extracts/literatura.json')
raksti_bizness = pd.read_json('extracts/finanses.json')
raksti_atputa = pd.read_json('extracts/atputa.json')
raksti_kriminalzinas = pd.read_json('extracts/kriminalzinas.json')
raksti_kino = pd.read_json('extracts/kino.json')
raksti_muzika = pd.read_json('extracts/muzika.json')
raksti_auto = pd.read_json('extracts/auto.json')
raksti_tehnologijas = pd.read_json('extracts/tehnologijas.json')
data_news = pd.concat([raksti_sports, raksti_politika,
    raksti_literatura, raksti_bizness,
    raksti_atputa, raksti_kriminalzinas,
    raksti_kino, raksti_muzika,
    raksti_auto, raksti_tehnologijas])
data_news.head()
print(data_news.category.value_counts())
print(data_news.shape)
```

```python
import matplotlib.pyplot as plt

def word_count(text):
    return len(str(text).split(' '))

data_news['word_count'] = data_news['body'].apply(word_count)

labels = ['Atpūta', 'Auto', 'Finanses', 'Kino', 'Kriminālziņas', 'Literatūra', 'Mūzika', 'Politika', 'Sports', 'Tehnoloģijas']

for label in labels:
    data_news_current = data_news.loc[data_news['category'] == label] 
    xlabel = "Vārdu skaits vidēji - " + str(round(data_news_current['word_count'].mean(), 2)) + ", max - " + str(data_news_current['word_count'].max())
    plt.xlabel(xlabel)
    plt.ylabel("Rakstu skaits")
    plt.hist(data_news_current['word_count'], bins = 100, color = 'red')
    plt.title(label, fontsize = 13)
    plt.xlim(0, 3000)
    plt.show()
```
